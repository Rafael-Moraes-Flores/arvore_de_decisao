```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Vamos supor que 'df' é o seu DataFrame com os dados
# Selecione apenas as colunas numéricas relevantes para calcular a correlação
data_numeric = df

# Calcule a matriz de correlação
correlation_matrix = data_numeric.corr()

# Crie um heatmap de correlação usando seaborn
plt.figure(figsize=(8, 6))  # Tamanho opcional da figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap de Correlação")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
df['area'] = pd.factorize( df['hsc_s'] )[0]
df['sexo'] = pd.factorize( df['gender'] )[0]
df['experiencia'] = pd.factorize( df['workex'] )[0]
```

```python
x = df[['area', 'sexo', 'experiencia']]
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=18)
```

```python
from sklearn.tree import DecisionTreeClassifier
arvore = DecisionTreeClassifier()
arvore.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

predicao = arvore.predict(X_test)
print(confusion_matrix(y_test, predicao))
print(classification_report(y_test, predicao))
```

```python
predicao = arvore.predict(X_test)
print(confusion_matrix(y_test,predicao))
print(classification_report(y_test,predicao))
```

```python
from sklearn import tree
features = list(x.columns)
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(20,20))
tree.plot_tree(arvore, feature_names=features, class_names=arvore.classes_)
```
